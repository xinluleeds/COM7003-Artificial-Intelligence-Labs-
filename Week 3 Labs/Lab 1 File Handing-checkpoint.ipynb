```python
# Type conversion
num = 10
num_str = str(num)  # Convert int to string
print(num_str, type(num_str))

str_to_int = int("123")  # Convert string to int
print(str_to_int, type(str_to_int))

string_to_list = list("Python")  # Convert string to list
print(string_to_list, type(string_to_list))


```

```python
# Numeric operations
x = 15
y = 4
print("Addition:", x + y)
print("Division:", x / y)
print("Floor Division:", x // y)

# String operations
greeting = "Hello"
name = "Python"
print(greeting + ", " + name + "!")

# List and set operations
my_list = [1, 2, 3]
my_list.append(4)
print("Modified list:", my_list)

my_set = {1, 2, 3}
my_set.add(4)
print("Modified set:", my_set)
```

```python
# Create a dictionary
person = {"name": "Alice", "age": 25, "city": "New York"}
print(person)

# Access a value
print("Name:", person["name"])

# Modify a value
person["age"] = 26
print("Updated age:", person["age"])

# Add a new key-value pair
person["profession"] = "Engineer"
print("Updated dictionary:", person)

# Delete a key-value pair
del person["city"]
print("After deletion:", person)
```

```python
# Boolean variables
is_active = True
is_logged_in = False

# Logical operators
print("AND:", is_active and is_logged_in)
print("OR:", is_active or is_logged_in)
print("NOT:", not is_active)

# Comparison operators
x = 10
y = 20
print("x > y:", x > y)
print("x == y:", x == y)
```

```python
# Collect user data
name = input("Enter your name: ")
age = int(input("Enter your age: "))
hobbies = input("Enter your hobbies (comma-separated): ").split(",")

# Store in dictionary
user_data = {"name": name, "age": age, "hobbies": hobbies}

# Display data
print("\nUser Information:")
print(f"Name: {user_data['name']}")
print(f"Age: {user_data['age']}")
print(f"Hobbies: {', '.join(user_data['hobbies'])}")
```
