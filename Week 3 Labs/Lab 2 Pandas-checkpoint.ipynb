```python
import pandas as pd

# Define the file path
file_path = r'C:\Users\X.Lu\COM7003 AI Labs\Week 3 Labs\shopping.csv'

# Read the CSV file
df = pd.read_csv(file_path)


# (Optional) Modify the data if needed
# Example: Adding a new column
df['discounted_price'] = df['price'] * 0.9  # Applying a 10% discount

# Write the updated DataFrame back to the same file
df.to_csv(file_path, index=False)

print(f"Data has been successfully written back to {file_path}.")
print("First 5 rows of the original file:")
print(df.head())
```
